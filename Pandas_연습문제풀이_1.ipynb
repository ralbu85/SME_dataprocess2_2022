```python
import numpy as np
low = np.quantile(df['Income'],0.25) #소득하위 25
high = np.quantile(df['Income'],0.75) #소득상위 25
print(low, high)
```

```python
# 소득 하위 25% 이하에 해당하는 사람들
low_bal=df[df['Income']<low]['Balance'].mean()
high_bal=df.query(f"Income>{high}")['Balance'].mean()
print(low_bal, high_bal)
```

## 연습문제 2-3

```python
def ageProc(x):
    if x<30: # 30 미만이면
        return 'Young'
    elif x<60: # 그렇지 않고 60미만이면(30~60)
        return 'Mature'
    else: # 그렇지 않다면 (60이상)
        return 'Old'

df['AgeCategory']=df['Age'].apply(ageProc)
df
```

## 연습문제 2-4

```python
def asianProc(x): # 행전체가 들어온다고 가정
    if x['Ethnicity'] == 'Asian':
        return x['Rating']+50
    else:
        return x['Rating']-50

# asianProc(df.loc[0])
df['newRating']=df.apply(asianProc, axis=1)
df
```
